```python
# Creating variables for initial statte

initial_bays = [[5],[1,6],[4],[3,2]]
initial_crane_position = 3
initial_crane_container_held = None
initial_cost = 0
heavy_crates = [1,3,6]


# Function for printing current state

def print_state(state_and_time):
  state = state_and_time[0]
  for i in range(len(state[0])):
    if i == state[1] :
      print(str(state[0][i])+" <-["+str(state[2])+"]-")
    else:
      print(str(state[0][i]))
  print("Total cost: "+str(state_and_time[1]))
  print()


initial_state = ([initial_bays,initial_crane_position,initial_crane_container_held,heavy_crates], 0)

print_state(initial_state)
```

```python
import copy

# Function to compute next state

def perform_action(state,action):
  new_state = copy.deepcopy(state[0])
  new_state_cost = state[1]
  cost = 0
  if action == "RIGHT":
    if new_state[1] >= (len(new_state[0]) -1):
      return None
    new_state[1] = int(new_state[1])+1
    if new_state[2] in new_state[3]:
      cost = 5
    else:
      cost = 3
  if action == "LEFT":
    if new_state[1] <= 0:
      return None
    new_state[1] = int(new_state[1])-1
    if new_state[2] in new_state[3]:
       cost = 5
    else:
       cost = 3
  if action == "DROP":
    if new_state[2] == None or len(new_state[0][new_state[1]]) >= 3:
      return None
    container = new_state[2]
    new_state[2] = None
    new_state[0][new_state[1]].append(container)
    cost = 2
  if action == "PICK":
    if new_state[2] != None:
      return None
    if len(new_state[0][new_state[1]]) == 0:
      return None
    container = new_state[0][new_state[1]].pop()
    new_state[2] = container
    cost = 4 - len(new_state[0][new_state[1]])
  return new_state, new_state_cost+cost

def perform_action_sequence(state,actions):
  new_state = state
  for action in actions:
    new_state = perform_action(new_state, action)
  return new_state
```

```python
import time
import numpy as np

all_actions = ["DROP", "PICK", "RIGHT", "LEFT"]

# function used for checking if action can actually be performed by crane
def is_action_valid(state, action):
    if perform_action(state, action):
      return True
    else:
      return False
```

```python
# Defining funcition to check goal state
# (Crates 1, 2, and 3 are placed in bay 2 in ascending order)
# (The crane finishes above an empty bay)

def is_goal_state(state):
  if state[0][1] == [1,2,3] and state[0][state[1]] == []:
    return True
```

```python
# Defining function to check if state is unique

def is_unique_state(state, unique_states):
    if state in unique_states:
        return False
    return True
```

```python
# Defining Breadth First Search

def bfs(initial_state, possible_actions = all_actions):
    frontier = []
    frontier.append((copy.deepcopy(initial_state),[]))
    unique_states = []
    unique_states.append((copy.deepcopy(initial_state[0])))

    start = time.time()

    while frontier:
        state = frontier.pop(0)

        if is_goal_state(state[0][0]):
            return state
        for action in possible_actions:
            if is_action_valid(state[0],action):
                # apply the action
                new_state = perform_action(state[0],action)
                # check if new state is unique or has been reached before
                if is_unique_state(new_state[0], unique_states):
                    new_actions = state[1].copy()
                    new_actions.append(action)
                    # add the new state in the frontier
                    frontier.append((new_state, new_actions))
                    unique_states.append(new_state[0])




        end = time.time()
        if end-start > 20:
            raise TimeoutError("Execution is taking too long to terminate.")
```

```python

print_state(initial_state)
solution = bfs((initial_state))
print(str(solution[1]))
print_state(solution[0])
print("Steps Taken: "+str(len(solution[1])))
```
